## Analysis 2 : Predictive modelling

### Introduction

The result of the first analysis shows that the accidents is significantly reduced on the weekend, and concentrated on Tuesday to Friday from March to December，so the group infered that the date may affect the number of accidents. In addition, weather conditions may also affect the number of accidents because severe weather conditions are more likely to cause traffic accidents. This report will constructe three machine-learning models (LinearRegression, Support vector machine and Random Forest) to predict the numbers of accidents in the future based on the date and weather conditions.

### Dataset
All the added data (Source:http://www.bom.gov.au/climate/data-services/station-data.shtml) used in this part is as follows:

- The daily weather data from 2013 to 2018: *Weather_Condition.csv* 

- The forcasted weather data from 16/6/2020 to 25/6/2020: *future_weather.csv*

### Workflows

#### Step 1: Data-Preprocessing
Firstly, the group checked the collected dataset and find there are some missing value, thus we filled the missing value with the median. Next, for the float variables from attributes temperature, rainfall and solar exposure, the group applied the normalization on them by Max-Min Scale. Finally, the variables from attributes months and days are discontinuous variable which need to be encoded. There are two common encode methods: Ordinal Encode and One-Hot Encode, therefore, this report create two function to implement different encode methods for these discontinuous variables and will compare their effect in the validation part.

```python
t3_2 = ap.Accident_Predict()
```

```python
x_train,y,x_predict = t3_2.get_train_data(acc_all_year)
```

```python
# this cell is to check 
x_train.isnull().any() 
```

```python
x_filled = t3_2.fill_missing_value(x_train,'Solar_exposure')
```

```python
x_train_scale,x_predict_scale = t3_2.normalization(x_filled,x_predict)
```

```python
x_train_ordEnc,x_predict_ordEnc = t3_2.ordinalEncode(x_train_scale,x_predict_scale)
```

```python
x_train_oneHot,x_predict_oneHot = t3_2.oneHotEncode(x_train_scale,x_predict_scale)
```

#### Step 2: Cross-Validation
In this Part, the group utilize 10 folds cross-validation to evaluate the models with different encoded dataset, we set the mean R-square as our evaluation standard to judge the goodness of fit for a model.

```python
cross_val = t3_2.model_validation(x_train_ordEnc,x_train_oneHot,y)
```

```python
cross_val
```

The result shows that dataset with One-Hot Encode perform better than ordinal, and the Random Forest model get best R-square value among the tested models.
